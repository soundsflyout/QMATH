# Lecture 2: Complex numbers

By and large, we will be interested in pairs of complex numbers, since these are the objects that model a single quantum bit of information. Because of this, we will invest today in teaching the basics of complex numbers.

What does it mean to say complex numbers are *numbers?* We will not go into great detail on this front, but like vectors, a *number field* can be abstractly characterized as those classes $F$ for which
- The integers $0,1,2,\ldots$ belong into $F$.
- We can extend our notion of $+$ on $0,1,2,\ldots$ to all of $F$ so that the integer $0$ is our 'zero element': $x+0 = x$ for any $x \in F$. It's 'commutative' and 'associative' as usual.
- As well, we have an inverse $-$ so that for every $x$, there is a $-x$ that satisfies $x+(-x) = 0$.
- We can extend our notion of $\cdot$ on $0,1,2,\ldots$ so that the integer $1$ is our 'one': $x \cdot 1 = x$ for all $x \in F$. It's 'commutative' and 'associative'.
- We also have an inverse ${}^{-1}$ as long as $x \neq 0$: $x^{-1}$ is so that $x \cdot x^{-1} = 1$).
- Finally, we have the usual distribution rule:
$$
\begin{align*}
    x\cdot(y+z) &= x\cdot y + x \cdot z \;.
\end{align*}
$$

We won't spend too much time working out the intricacies of number fields. The key takeaway that we should have is that number fields are things where you are allowed to $+,-,\cdot, {}^{-1}$. The abstract field class can be defined as follows:

```python
from abc import ABC, abstractmethod

class Field(ABC):
    @abstractmethod
    def __add__(self, other: Field) -> Field:
        pass

    @abstractmethod
    def __neg__(self) -> Field:
        """
        Gives the negative of a number
        :return: -self
        """
        pass

    @abstractmethod
    def __mul__(self, other: Field) -> Field:
        pass

    @abstractmethod
    def inv(self) -> Field:
        """
        Gives the multiplicative inverse of a number
        :return: 1/self
        """
        pass

    def __sub__(self, other: Field) -> Field:
        """
        Gives the difference of self and other.
        :return: self-other
        """
        return self+(-other)

    def __truediv__(self, other: Field) -> Field:
        """
        Gives the division of self and other.
        :return: self/other
        """
        return self*(other.inv())
```

## Non-example
The integers $\mathbb{Z} = \{\ldots, -3,-2,-1,0,1,2,3,\ldots\}$ with the usual $\cdot$ and $+$ is *not* a number field.

## Example 1
The rational numbers $\mathbb{Q} = \{\frac{a}{b}: a,b \in \mathbb{Z}, b \neq 0\}$ is a number field. We know that $\frac{a}{b}^{-1} = \frac{b}{a}$ as long as $a \neq 0$ and that our usual rules of $+$ and $\cdot$ apply.

## Example 2
The real line $\mathbb{R}$ of all numbers with arbitrarily long decimal expansions is a number field. We won't go into the precise definition of what 'arbitrarily long decimal expansions' are, but just notice that you can $+,-,\cdot,{}^{-1}$ any number.

## Example 3
This next example is the most important one. Recall from last time that we made the vector space
$$
    P_2(x,\mathbb{R}; x^2+1)
$$
of quadratic polynomials with the additional constraint that $x^2+1 = 0$. We saw that every quadratic reduces to a linear object: an object of the form $a+bx$, because we can always replace $x^2$ with $-1$.

We want to turn $P_2(x,\mathbb{R};x^2+1)$ into a number field. We already have an adequate $+,-$, so it is enough to construct a $\cdot$ that satisfies the distribution rule and show that every non-zero quadratic is invertible.

Of course, we have the usual notion of $\cdot$ on polynomials, which we know will satisfy the distribution rule. Let's calculate it on the reduced linear terms, along with using the relation $x^2 = -1$.
$$
\begin{align*}
    (a+bx)\cdot(c+dx) &= ac + (bc + ad)x + (bd)x^2 \\
    &= ac + (bc + ad)x (bd)(-1) \\
    &= (ac-bd) + (bc+ad)x \;.
\end{align*}
$$

This gives us a notion of $\cdot$. Suppose that $a+bx \neq 0$ (so $a,b$ are not both zero). The case when $b = 0$ is easy, since $(a+0x)^{-1} = 1/a$, so let's look at the case where $b \neq 0$. That is, let's look for numbers $c,d$ so that
$$
    (a+bx) \cdot(c+dx) = 1 \;.
$$
By our calculation, we know that this means that we need
$$
    (ac-bd) + (bc+ad)x = 1 \;.
$$
In other words, we want
$$
\begin{align*}
    ac-bd &= 1 \\
    bc+ad &= 0  \;.
\end{align*}
$$
Can we always find $c,d$ that do this? First of all, the second equation gives us the relation
$$
\begin{align*}
    bc &= -ad \\
    \frac{c}{d} &= \frac{-a}{b}
\end{align*}
$$
provided that the $d$ we choose is not zero. Dividing the first equation by $d$, we get
$$
\begin{align*}
    a\cdot\left(\frac{c}{d} \right) - b &= \frac{1}{d} \\
    a\cdot \left( \frac{-a}{b} \right) - b &= \frac{1}{d} \\
    \frac{-1}{b}\left(a^2 + b^2\right) &= \frac{1}{d} \;.
\end{align*}
$$
In other words, we have
$$
    d = \frac{-b}{a^2 + b^2} \;.
$$
We know that since $b \neq 0$, $a^2 +b^2 \neq 0$, and this is sensible and non-zero.

Using our second equation gives us
$$
    c = \frac{-a}{b} \cdot d = \frac{a}{a^2+b^2} \;.
$$
Putting everything together, we get the relation
$$
    (a+bx)^{-1} = \frac{a}{a^2+b^2} - \frac{b}{a^2+b^2}x \;.
$$

In order to distinguish the vector space $P_2(x,\mathbb{R};x^2+1)$ and this new class that has the additional $\cdot$ and ${}^{-1}$ rules, we denote this number field by $\mathbb{C}$ -- the field of complex numbers.

Mathematicians often use $i$ to denote the unknown in this number field, so that arbitrary objects in $\mathbb{C}$ look like $a+bi$. In python, they use the letter $j$ for the unknown: you'll see $a+bj$ instead.

The two parts $a$ and $b$ in $a+bi$ are called the 'real part' and the 'imaginary part' respectively.



We can code up the class of complex numbers by modifying our previous Quad class.

```python
# Class of Complex numbers
class Complex(Field):
    def __init__(self, real: float = 0, imag: float = 0, quadratic_: float = 0):
        self.real = real
        self.imag = imag
        self.quadratic_ = quadratic_


    def reduce_(self) -> Complex:
        """
            Reduces a polynomial to is linear form. This should never be called directly.
        """
        self.real += -self.quadratic_
        self.quadratic_ = 0
        return self

    def __add__(self, other: Complex) -> Complex:
        new = Complex()
        new.real = self.real + other.real
        new.imag = self.imag + other.imag
        new.quadratic_ = self.quadratic_ + other.quadratic_
        new.reduce_()
        return new

    def __neg__(self) -> Complex:
        return Complex(-self.real, -self.imag)

    def __mul__(self, other: Complex) -> Complex:
        new = Complex()
        new.real = self.real * other.real - self.imag * other.imag
        new.imag = self.imag * other.real + self.real * other.imag
        return new

    def inv(self) -> Complex:
        new = Complex()
        try:
            new.real = self.real/(self.real**2 + self.imag**2)
            new.imag = -self.imag/(self.real**2 + self.imag**2)
            return new
        except ZeroDivisionError:
            raise ZeroDivisionError('Division by zero')

    def __eq__(self, other: Complex) -> bool:
        """ Define equality to account for the fact that x^2 = -1 """
        a = self.reduce_()
        b = other.reduce_()
        return a.real == b.real and a.imag == b.imag

    def __ne__(self, other: Complex) -> bool:
        return not self.__eq__(other)


    # This is the value r in the polar representation
    def length(self):
        return np.sqrt(self.real**2 + self.imag**2)

    def angle(self):
        if self.real > 0:
            return np.arctan(self.imag/self.real)
        elif self.real == 0:
            if self.imag >= 0:
                return np.pi/2
            else:
                return -np.pi/2
        else:
            if self.imag > 0:
                return np.pi - np.arctan(-self.imag/self.real)
            else:
                return -np.pi - np.arctan(-self.imag/self.real)

    def conj(self):
        return Complex(self.real, -self.imag)

    def __str__(self):
        return f'{self.real}+{self.imag}i'

    @staticmethod
    def exp(z: Complex) -> Complex:
        return Complex(real=np.exp(z.real))*Complex(np.sin(z.imag), np.cos(z.imag))

    def plot(self, color: str = None) -> None:
        """
        Plots the complex number a+bi as a pair $(a,b)$ using seaborn
        """
        sns.lineplot(x=[0, self.real], y=[0, self.imag], color=color)

```

# The complex conjugate
The formula for an inverse is quite intricate, and in fact there is a nice way of interpreting what is going on if we use our 'vectors as line segments' picture for $\mathbb{C}$.

Since any element $a+bi$ of $\mathbb{C}$ is mechanically the same as the pair $(a,b)$ as a vector space, we can draw it as a line segment:

```python
points =[Complex(real=3, imag=2), Complex(real=-1, imag=3), Complex(real=2, imag=-3), Complex(real=-2, imag=-4)]

x_min = min([pt.real for pt in points]+[0])
x_max = max([pt.real for pt in points]+[0])
y_min = min([pt.imag for pt in points]+[0])
y_max = max([pt.imag for pt in points]+[0])

for pt in points:
    pt.plot()

for y in range(y_min, y_max+1):
    sns.lineplot(x=[x_min, x_max], y=[y,y], c='gray', alpha=0.2)
for x in range(x_min, x_max+1):
    sns.lineplot(x=[x,x], y=[y_min, y_max], c='gray', alpha=1)
plt.title(f'Plot of points {[str(x) for x in points]}')
plt.show()
```

One of the complicated terms in $z^{-1}$ is the $1/(a^2+b^2)$ term. Let's make all the line segments have length $1$ (that is, $\sqrt{a^2+b^2} = 1$) and picture some vectors and their inverses. Below are four vectors and their inverses, where a vector and its inverse are given the same colour.

```python
import math
points =[Complex(real=3, imag=2), Complex(real=-1, imag=3), Complex(real=2, imag=-3), Complex(real=-2, imag=-4)]
colors = ['red', 'green', 'blue', 'orange']


x_min = min([pt.real for pt in points]+[0])
x_max = max([pt.real for pt in points]+[0])
y_min = min([pt.imag for pt in points]+[0])
y_max = max([pt.imag for pt in points]+[0])

for idx, pt in enumerate(points):
    # First let's normalize the points
    pt.real, pt.imag = pt.real/pt.length(), pt.imag/pt.length()
    inverse_pt = pt.inv()
    pt.plot(color=colors[idx])
    inverse_pt.plot(color=colors[idx])

sns.lineplot(x=[-1,1], y=[0,0], c='gray', alpha=0.2)
sns.lineplot(x=[0,0], y=[-1,1], c='gray', alpha=1)
plt.xlabel('real')
plt.ylabel('imaginary')
plt.show()
```

**Definition** Given a complex number $z = a+bi$, its *complex conjugate* is the number $\bar{z} = a-bi$.

## A taste of what's to come: the Bloch sphere
At the beginning, I mentioned that a quantum bit for us will often be stored as a pair of complex numbers. Roughly, a quantum bit will store $(1,0)$ as a '1' and $(0,1)$ as a $'0'$, and all the numbers $(z,w)$ in between will represent some value that behaves sometimes like '1' and sometimes like '0'.

A pair of complex numbers is a four dimensional beast, and we cannot possibly hope to picture such a thing in a productive way. However, we are saved due to two facts associated to models of quantum bits:

1. Quantum bits must always have length $1$. That is, if $(z,w)$ represents a quantum bit, then $|z|^2 + |w|^2 = 1$.
2. There is no detectable difference between the quantum bit $(z,w)$ and the quantum bit $e^{it} \cdot (z,w) = (e^{it}\cdot z, e^{it} \cdot w)$. So, we can consider them as describing the same object. Let's denote this fact by the symbol $\equiv$. That is, we say $(z,w) \equiv e^{it}\cdot(z,w)$ for any number $t \in \mathbb{R}$.

Once we apply these two additional restrictions, we shall see that we can actually represent a quantum bit as a point on a sphere with radius $1$. This is called the **Bloch sphere** representation. It is not very intuitive at the beginning, so I introduce it now to ease us into thinking about it. Very roughly, the Bloch sphere will be so that
- every point on the sphere represents a unique quantum bit,
- the north pole $(1,0,0)$ will represent the quantum bit $(1,0)$, and
- the south pole $(-1,0,0)$ will represent the quantum bit $(0,1)$.

First of all, we start by writing our pair of complex numbers in polar form: say $(z,w) = (r_1 \cdot e^{it_1}, r_2 \cdot e^{it_2})$. Observe that $(z,w) = e^{it_1}\cdot(r_1, r_2 \cdot e^{i(t_2-t_1)})$ by how products of complex exponentials behave. By rule 2, we can take $\phi = t_2 - t_1$, and say that $(z,w) \equiv (r_1, r_2 \cdot e^{i \phi})$.

Rule 1 tells us that $r_1^2 + r_2^2 = 1$. By properties of cosines and sines, we know that there is a number $\theta$ so that $r_1 = \cos(\theta/2)$ and $r_2 = \sin(\theta/2)$ (I explain why we choose $\theta/2$ instead of $\theta$ soon. In fact, because $r_1, r_2 \geq 0$, we know that $\theta \in [0,\pi]$. This means that $(z,w) \equiv (\cos(\theta), \sin(\theta) \cdot e^{i \phi})$.

Finally, we use the definition of complex exponentials to write $e^{i \phi}$ as $\cos(\phi) + \sin(\phi)i$, giving us
$$
    (z,w) \equiv (\cos(\theta/2), \sin(\theta/2)\cos(\phi) + \sin(\theta/2)\sin(\phi) i) \;.
$$
Since the pair $\cos(\phi) + \sin(\phi)i$ geometrically the same as the pair $(\cos(\phi), \sin(\phi))$, we get a geometric representation
$$
    (z,w) \sim (\cos(\theta/2), \sin(\theta/2)\cos(\phi), \sin(\theta/2)\sin(\phi)) \;.
$$
Now, looking at values of $\theta \in [0,\pi]$ and $\phi \in (-\pi,\pi]$ will only give us a hemisphere. In order for the north and south poles to represent the quantum bits $(1,0)$ and $(0,1)$, we need to stretch $\theta$ by $2$. That is, the Bloch sphere representation of $(z,w)$ is
$$
    Bloch\_sphere(z,w) = (\cos(\theta), \sin(\theta)\cos(\phi), \sin(\theta)\sin(\phi)) \;.
$$

Conversely, given a point on the sphere $(\cos(\theta), \sin(\theta)\cos(\phi), \sin(\theta)\sin(\phi))$, its corresponding quantum bit is
$$
    (z,w) = (\cos(\theta/2), \sin(\theta/2)e^{i \phi}) \;.
$$


```python
from qiskit.visualization import plot_bloch_vector
from ipywidgets import interact, FloatSlider
theta = FloatSlider(value=np.pi/2,
                    min=0,
                    max=np.pi,
                    step=0.1,
                    disabled=False,
                    continuous_update=False,
                    orientation='horizontal',
                    readout=True,
                    readout_format='.1f')
phi = FloatSlider(value=0,
                    min=0,
                    max=2*np.pi,
                    step=0.1,
                    disabled=False,
                    continuous_update=False,
                    orientation='horizontal',
                    readout=True,
                    readout_format='.1f')

def plot_bloch(theta, phi):
    return plot_bloch_vector([1,theta, phi], coord_type = 'spherical')

interact(plot_bloch, theta=theta, phi=phi);
```

Let's check that the north and south poles really do correspond to the quantum bits $(1,0)$ and $(0,1)$.
